```python
import json
import os
import random
import warnings

import numpy as np
import pandas as pd
import torch
from tqdm import tqdm, trange

warnings.filterwarnings("ignore")

# Set up seeds
seed = 42
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)
torch.backends.cudnn.deterministic = True
random.seed(seed)


def read_jsonl_file(jsonl_path):
    with open(jsonl_path, "r") as f:
        for line in f:
            yield json.loads(line)
```

```python
path = "defects4j.jsonl"


# convert to csv
data = []
for d in read_jsonl_file(path):
    data.append(d)


df = pd.DataFrame(data)
#bug_id,func_before,func_after
df.rename(columns={"buggy_code": "func_before", "fixed_code": "func_after"}, inplace=True)

df.to_csv("defects4j.csv", index=False)
```
